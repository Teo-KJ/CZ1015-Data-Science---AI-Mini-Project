```python
f, axes = plt.subplots(2,1,figsize=(15,10))

# Top 15 happiest countries
topCountry=d2.sort_values(by=['Happiness score'],ascending=False).head(n=15)
sb.barplot(x='Country',y='Happiness score', data=topCountry, palette = 'Set2', ax=axes[0])
axes[0].set(xlabel='Country', ylabel='Happiness score')
tl = axes[0].get_xticklabels()
axes[0].set_xticklabels(tl, rotation=45);

# Top 15 unhappiest countries
worstCountry=d2.sort_values(by=['Happiness score'],ascending=True).head(n=15)
sb.barplot(x='Country',y='Happiness score', data=worstCountry, palette = 'Set2', ax=axes[1])
axes[1].set(xlabel='Country', ylabel='Happiness score')
tl = axes[1].get_xticklabels()
axes[1].set_xticklabels(tl, rotation=45);
```

```python
d2_subplots = d2[['Country','Happiness score','Whisker-high','Whisker-low','Dystopia (1.92) + residual','Explained by: GDP per capita','Explained by: Social support','Explained by: Healthy life expectancy','Explained by: Freedom to make life choices','Explained by: Generosity', "Explained by: Perceptions of corruption"]]

d2_subplots.plot(subplots=True, figsize=(12, 16))
d2_subplots.hist(edgecolor = 'white', linewidth = 1, figsize = (20,16))
plt.show()
```
