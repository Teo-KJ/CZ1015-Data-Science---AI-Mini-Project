```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import math
import matplotlib.pyplot as plt
```

```python
d1 = pd.read_excel("Data/World Happiness Report/WHR2018.xlsx", sheet_name='Table2.1')
d2 = pd.read_excel("Data/World Happiness Report/WHR2018.xlsx", sheet_name='Figure2.2')
d3 = pd.read_excel("Data/World Happiness Report\WHR2018.xlsx", sheet_name='Figure2.3')
d4 = pd.read_excel("Data/World Happiness Report/WHR2018.xlsx", sheet_name='Figure2.4')
d5 = pd.read_excel("Data/World Happiness Report/WHR2018.xlsx", sheet_name='SupportingFactors')
```

```python
print(d1.shape)
d1 = d1.rename(columns={'country':'Country'})
# d1['Average'] = d1.groupby('Country')['Life Ladder'].transform('mean')
# d1['Rank'] = d1['Average'].rank(method='dense', ascending=False)
d1.head()
```

```python
print(d2.shape)
d2 = d2.iloc[0:157, 0:11].sort_values(by=['Country']).reset_index(drop=True)
d2.head()
```

```python
print(d3.shape)
d3= d3.sort_values(by=['Country']).reset_index(drop=True)
d3.head()
```

```python
print(d4.shape)
d4= d4.sort_values(by=['Country']).reset_index(drop=True)
d4.head()
```

```python
print(d5.shape)
d5 = d5.sort_values(by=['country']).reset_index(drop=True)
d5 = d5.rename(columns={'country':'Country'})
d5.head()
```

```python
f, axes = plt.subplots(2,1,figsize=(15,10))

# Top 15 happiest countries
topCountry=d2.sort_values(by=['Happiness score'],ascending=False).head(n=15)
sb.barplot(x='Country',y='Happiness score', data=topCountry, palette = 'Set2', ax=axes[0])
axes[0].set(xlabel='Country', ylabel='Happiness score')
tl = axes[0].get_xticklabels()
axes[0].set_xticklabels(tl, rotation=30);
axes[0].set_ylim(6.8, 8) # Truncate y-axis to show greater change

# Top 15 unhappiest countries
worstCountry=d2.sort_values(by=['Happiness score'],ascending=True).head(n=15)
sb.barplot(x='Country',y='Happiness score', data=worstCountry, palette = 'Set2', ax=axes[1])
axes[1].set(xlabel='Country', ylabel='Happiness score')
tl = axes[1].get_xticklabels()
axes[1].set_xticklabels(tl, rotation=30);
axes[1].set_ylim(2.8, 4); # Truncate y-axis to show greater change
```

```python
data = pd.merge(d5, d2, on='Country')
data = pd.merge(data, d4, on='Country')
data = pd.merge(data, d3, on='Country')

# newcol_Average = data.groupby('Country')['Happiness score'].transform('mean')
# data.insert(loc=2, column='Average', value=newcol_Average)

newcol_Rank = data['Happiness score'].rank(method='dense', ascending=False)
data.insert(loc=2, column='Rank', value=newcol_Rank)

print(data.columns)
data.columns
```

```python
data = data.sort_values(by=['Rank'], ascending=True).fillna(0).reset_index(drop=True)
data
```

```python
data.hist(edgecolor = 'white', linewidth = 1, figsize = (20,16))
plt.show()
```

```python
df = data[[
       'Rank',
       'Life ladder, 2015-2017',
       'Log of GDP per person, 2015-2017', 
       'GDP per person, 2015-2017',
       'Healthy life expectancy, 2015-2017', 
       'Social support, 2015-2017',
       'Freedom to make life choices, 2015-2017',
       'Generosity, 2015-2017, without adjustment for GDP per person',
       'Perceptions of corruption, 2015-2017',
       'Happiness score',
       'Dystopia (1.92) + residual',
       'Average happiness of foreign born',
       'Average happiness of locally born'
        ]]
df.plot(subplots=True, figsize=(12, 16));
```

```python
plt.figure(figsize=(16,14))
# corr = data.drop(['Country','Region indicator','Rank'],axis = 1).corr()
corr = df.corr()
sb.heatmap(corr, cbar = True, square = True, annot=True, linewidths = .5, fmt='.2f',annot_kws={'size': 10}) 
plt.show()
```

```python
sb.kdeplot(df['Happiness score'], df['GDP per person, 2015-2017'], shade=True)
plt.scatter(df['Happiness score'], df['GDP per person, 2015-2017'], alpha=0.2, color='green')
plt.xlabel('Happiness Score')
plt.ylabel('GDP per Capita')
plt.title('Happiness vs. GDP')
plt.show()

sb.kdeplot(df['Happiness score'], df['Healthy life expectancy, 2015-2017'], shade=True)
plt.scatter(df['Happiness score'], df['Healthy life expectancy, 2015-2017'], alpha=0.2, color='green')
plt.xlabel('Happiness Score')
plt.ylabel('Health (Life Expectancy)')
plt.title('Health (Life Expectancy) vs. GDP')
plt.show()
```

```python
plt.figure(figsize=(12,12))
sb.jointplot(x = 'GDP per person, 2015-2017', y = 'Happiness score', data = df, size=10, color='red')
plt.ylabel('Happiness score')
plt.xlabel('GDP per person, 2015-2017')
plt.title('GDP per person, 2015-2017 Vs Happiness')
plt.show()
```
